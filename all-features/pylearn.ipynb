## _WARNING: exploratory, not functional_

```python
import pandas
import sklearn.linear_model
import sklearn.preprocessing
import sklearn.metrics
```

```python
dwpc_mat_df = pandas.read_table('data/matrix/hetio-ind/DWPC-spread.tsv.bz2')
dwpc_mat_df.head(2)
```

```python
features = ['compound_treats', 'disease_treats'] + list(dwpc_mat_df.columns[7:])
X = dwpc_mat_df[features]
scaler = sklearn.preprocessing.StandardScaler()
X_standard = scaler.fit_transform(X)
y = dwpc_mat_df['status']
w = [1.0] * len(y)
```

```python
# L1 for lasso
model = sklearn.linear_model.LogisticRegression(penalty='l1')
model = model.fit(X_standard, y, sample_weight=w)
```

```python
# L1 for lasso
model = sklearn.linear_model.LogisticRegressionCV(penalty='l1', n_jobs=10, solver='liblinear', random_state=0, scoring=sklearn.metrics.roc_auc_score, verbose=1)
model = model.fit(X_standard, y, sample_weight=w)
```
